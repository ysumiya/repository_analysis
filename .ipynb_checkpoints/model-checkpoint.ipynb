```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

#  データセット

```python
df = pd.read_csv('./UCP_Dataset.csv', header=0, index_col=0, sep=';')
df
```

# 特徴量・教師データに使用する属性を指定

```python
X_columns = ['Simple UC', 'Average UC', 'Complex UC']
Y_name = 'Real_Effort_Person_Hours'
```

# 各属性と"Real_Effort_Person_Hours"の相関を見る．

## Simple UC

```python
%matplotlib inline
X_name = 'Simple UC'
data_X = df.loc[:, X_name].as_matrix()
data_Y = df.loc[:, Y_name].as_matrix()
plt.scatter(data_X, data_Y)
plt.xlabel(X_name)
plt.ylabel(Y_name)
```

## Average UC

```python
%matplotlib inline
X_name = 'Average UC'
data_X = df.loc[:, X_name].as_matrix()
data_Y = df.loc[:, Y_name].as_matrix()
plt.scatter(data_X, data_Y)
plt.xlabel(X_name)
plt.ylabel(Y_name)
```

## Complex UC

```python
%matplotlib inline
X_name = 'Complex UC'
data_X = df.loc[:, X_name].as_matrix()
data_Y = df.loc[:, Y_name].as_matrix()
plt.scatter(data_X, data_Y)
plt.xlabel(X_name)
plt.ylabel(Y_name)
```

## UUCW

```python
%matplotlib inline
X_name = 'UUCW'
data_X = df.loc[:, X_name].as_matrix()
data_Y = df.loc[:, Y_name].as_matrix()
plt.scatter(data_X, data_Y)
plt.xlabel(X_name)
plt.ylabel(Y_name)
```

# トレーニングデータとテストデータに分割

```python
X = np.empty((71,0), dtype='float64')
Y = df.loc[:, Y_name].as_matrix()

for col in X_columns:
    X = np.append(X, df.loc[:,col].as_matrix().reshape(-1,1), axis=1)
    
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y)

X_train
```

# 重回帰モデルで予測

```python
clf = linear_model.LinearRegression()
clf.fit(X_train, Y_train)
predict = clf.predict(X_test)
print predict
print Y_test
print('L1 norm : {}'.format(np.mean(abs((predict - Y_test)))))
```
